```python
import tensorflow as tf
print(tf.__version__)

from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
import re
```

```python
X = list()
Y = list()
X = [x+3 for x in range(-2, 43, 3)]

for i in X:
    output_vector = list()
    output_vector.append(i+1)
    output_vector.append(i+2)
    Y.append(output_vector)

print(X)
print(Y)
```

```python
X = np.array(X).reshape(15, 1, 1)
Y = np.array(Y)
print(Y)
```

```python
tf.keras.backend.clear_session()
model = Sequential()
LSTM
model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
model.add(Dense(2))
#RNN
#model.add(SimpleRNN(50, activation='tanh', input_shape=(1, 1)))
#model.add(Dense(2))

output_size = 2


model.compile(optimizer='adam', loss='mse')

model.fit(X, Y, epochs=1000, validation_split=0.2, batch_size=3)
```

```python
test_input = np.array([10])
test_input = test_input.reshape((1, 1, 1))

test_output = model.predict(test_input, verbose=0)
print(test_output)
```
